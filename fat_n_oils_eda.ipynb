```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('fatandoils.csv')

#df['fatoil_category'].value_counts()

#saturated and trans fat bad, mono and poly unsaturated good
```

```python
#foods with highest trans fats: Margarine-like shortening has the most along with any other margarine products, also a few salad dressingand and mayo
transfats = df['trans_fat(g)'].groupby(df['fatoil_category']).mean().sort_values(ascending=True)

#foods with highest saturated fats:
satfats = df['saturated_fat(g)'].groupby(df['fatoil_category']).mean().sort_values(ascending=True)

#foods with highest monounstaturated fats (good fats): 
monounsatfat = df['monounsat_fat(g)'].groupby(df['fatoil_category']).mean().sort_values(ascending=True)

#foods with highest polyunstaturated fats (good fats): 
polyunsatfat = df['polyunsat_fat(g)'].groupby(df['fatoil_category']).mean().sort_values(ascending=True)

#foods with highest cholesterol: Fish oil by a lot!! Butter fat and animal fat come next
cholesterol = df['cholesterol(mg)'].groupby(df['fatoil_category']).mean().sort_values(ascending=True)

#foods with highest sodium: Butter replacement, dressing, salad dressing, sandwhich spread
sodium = df['sodium(mg)'].groupby(df['fatoil_category']).mean().sort_values(ascending=True)

#foods with highest protein: Butter is the only thing with more than 2 grams of protein
protein = df['protein(g)'].groupby(df['fatoil_category']).mean().sort_values(ascending=True)

#foods with highest Kcal: Fishoil by a lot with shortening products following
kilojules = df['energy(Kj)'].groupby(df['fatoil_category']).mean().sort_values(ascending=True)
```

**Bad Fats**

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=True)

transfats.plot(kind="barh", ax=axes[0], color="blue")
axes[0].set_title("Avg Trans Fat by Category")
axes[0].set_xlabel("Trans Fats")
axes[0].set_ylabel("Category")

satfats.plot(kind="barh", ax=axes[1], color="red")
axes[1].set_title("Avg Saturated Fat by Category")
axes[1].set_xlabel("Saturated Fats")
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()
```

**Good Fats**

```python
fig, axes = plt.subplots(1,2, figsize=(12,4),sharey=True)

monounsatfat.plot(kind='barh',ax=axes[0], color="blue")
axes[0].set_xlabel("category")
axes[0].set_ylabel("Monounsaturated Fat")
axes[0].set_title('Avg Monounsat Fat by Fat/Oil Category')


polyunsatfat.plot(kind='barh',ax=axes[1], color="red")
axes[1].set_ylabel("category")
axes[1].set_xlabel("Polyunsaturated Fat")
axes[1].set_title('Avg Polyunsat Fat by Fat/Oil Category')

plt.tight_layout()
plt.show()
```

**Other variables**

```python
kilojules.plot(kind='barh')
plt.ylabel("category")
plt.xlabel("Kilojules")
plt.title('Avg KJs by Fat/Oil Category')
plt.show()
```
